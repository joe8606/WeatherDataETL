```python
import pandas as pd
```

```python
df = pd.read_csv(r'D:\即時值查詢 (4).csv', encoding='big5')
```

```python
import numpy as np
# 指定要轉換為數字的欄位範圍
columns_to_convert = [str(i) for i in range(24)]

# 將 '*' 替換為 NaN
df[columns_to_convert] = df[columns_to_convert].replace('*', np.nan)

# 將 'x' 替換為 NaN
df[columns_to_convert] = df[columns_to_convert].replace('x', np.nan)

# 將 '#' 替換為 NaN
df[columns_to_convert] = df[columns_to_convert].replace('#', np.nan)

# 將指定範圍的欄位從字串轉換為浮點數
df[columns_to_convert] = df[columns_to_convert].astype(float)

df['日期'] = pd.to_datetime(df['日期'])
```

```python
# 創建一個新的 'daily_average' 欄位，並計算平均值（忽略遺失值）
df['daily_average'] = df.iloc[:, 3:27].mean(axis=1, skipna=True)
```

```python
# 使用 unique() 函數獲取所有唯一的測站值
unique_stations = df['測站'].unique()
```

```python
# 選取 column0 到 column23 的資料
selected_columns = df.iloc[:, 3:27]

# 找出每個索引中的最大值，並放入一個列表中
max_values_list = selected_columns.max(axis=1).tolist()

df['O3_daily_max'] = max_values_list
```

```python
# 找出每個索引中前八大的值
top_eight_largest = selected_columns.apply(lambda row: row.nlargest(8), axis=1)

# 計算每個索引中前八大值的平均
average_top_eight = top_eight_largest.mean(axis=1)

# 將平均值轉換為一個列表
average_top_eight_list = average_top_eight.tolist()

df['O3_average_top_eight'] = average_top_eight_list
```

```python
station_df = pd.DataFrame()
station_df['測站'] = df['測站'].unique()
for i in df['測項'].unique():
    lst = []
    for j in df['測站'].unique():
        # 過濾出測站為基隆且測項為CO的資料
        filtered_df = df[(df['測站'] == j) & (df['測項'] == i)]
        lst.append(filtered_df['daily_average'].mean())
    station_df[i] = lst
```
