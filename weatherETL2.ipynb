```python
import pandas as pd
import os

# 指定存放CSV檔案的資料夾路徑
folder_path = r'D:\weiyi\data_202201'

# 創建一個空的DataFrame，用於儲存合併後的數據
merged_data = pd.DataFrame()

# 定義地區代碼和地區名稱的映射字典
region_mapping = {
    '01': '北海道',
    '02': '東北',
    '03': '東北',
    '04': '東北',
    '05': '東北',
    '06': '東北',
    '07': '東北',
    '08': '關東',
    '09': '關東',
    '10': '關東',
    '11': '關東',
    '12': '關東',
    '13': '關東',
    '14': '關東',
    '15': '中部',
    '16': '中部',
    '17': '中部',
    '18': '中部',
    '19': '中部',
    '20': '中部',
    '21': '東海',
    '22': '東海',
    '23': '東海',
    '24': '東海',
    '25': '近畿',
    '26': '近畿',
    '27': '近畿',
    '28': '近畿',
    '29': '近畿',
    '30': '近畿',
    '31': '中國',
    '32': '中國',
    '33': '中國',
    '34': '中國',
    '35': '中國',
    '36': '中國',
    '37': '中國',
    '38': '中國',
    '39': '中國',
    '40': '九州',
    '41': '九州',
    '42': '九州',
    '43': '九州',
    '44': '九州',
    '45': '九州',
    '46': '九州',
    '47': '沖繩'
}

# 遍歷資料夾中的所有檔案
for filename in os.listdir(folder_path):
    if filename.endswith(".csv"):
        # 讀取CSV檔案
        file_path = os.path.join(folder_path, filename)
        df = pd.read_csv(file_path, encoding='latin1')
        
        # 從檔名中提取地區信息
        region_code = filename.split('_')[1]
        region_name = region_mapping.get(region_code, '未知地區')  # 如果地區代碼不存在於字典中，則設置為'未知地區'
        
        # 將地區信息加入DataFrame作為一個新的欄位
        df['Region'] = region_name
        
        df['filename'] = filename
        
        # 將讀取的數據合併到主數據框中
        merged_data = pd.concat([merged_data, df], ignore_index=True)

# 將合併後的數據寫入Excel檔案，並以地區為Sheet名稱分別存儲
excel_output_path = r'D:\weiyi\202201.xlsx'
with pd.ExcelWriter(excel_output_path, engine='xlsxwriter') as writer:
    for region_name in merged_data['Region'].unique():
        region_df = merged_data[merged_data['Region'] == region_name]
        region_df.to_excel(writer, sheet_name=region_name, index=False)

print(f'合併完成，結果已保存至 {excel_output_path}')
```
